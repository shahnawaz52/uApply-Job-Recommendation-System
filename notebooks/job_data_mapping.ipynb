```python
def create_job_mapping_dictionary(data, key_column, value_column):
    mapping_dict = {}

    for index, row in data.iterrows():
        key_value = row[key_column]
        value = row[value_column]

        if key_value not in mapping_dict:
            mapping_dict[key_value] = []
        mapping_dict[key_value].append(value)

    return mapping_dict

# Example usage for company_name
company_specialities_mapping = create_job_mapping_dictionary(company_specialities, 'company_id', 'speciality')
company_industry_mapping = create_job_mapping_dictionary(company_industries, 'company_id', 'industry')
company_name_mapping = create_job_mapping_dictionary(companies, 'company_id', 'name')
company_desc_mapping = create_job_mapping_dictionary(companies, 'company_id', 'description')
company_linkedin_url_mapping = create_job_mapping_dictionary(companies, 'company_id', 'url')
job_industry_mapping = create_job_mapping_dictionary(industries, 'industry_id', 'industry_name')
job_industries['industry_name'] = job_industries['industry_id'].map(job_industry_mapping)
job_industry_name_mapping = create_job_mapping_dictionary(job_industries, 'job_id', 'industry_name')
skills_mapping = create_job_mapping_dictionary(skills, 'skill_abr', 'skill_name')
job_skills['skill_name'] = job_skills['skill_abr'].map(skills_mapping)
job_skills_mapping = create_job_mapping_dictionary(job_skills, 'job_id', 'skill_name')
# Create a new column in job_postings based on the mapping
job_postings['company_speciality'] = job_postings['company_id'].map(company_specialities_mapping)
job_postings['company_industry'] = job_postings['company_id'].map(company_industry_mapping)
job_postings['company_name'] = job_postings['company_id'].map(company_name_mapping)
job_postings['company_description'] = job_postings['company_id'].map(company_desc_mapping)
job_postings['linkedin_url'] = job_postings['company_id'].map(company_linkedin_url_mapping)
job_postings['company_industry'] = job_postings['job_id'].map(job_industry_name_mapping)
job_postings['job_skills'] = job_postings['job_id'].map(job_skills_mapping)
```
